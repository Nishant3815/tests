```python
#Direct Substitution - Good Convergence
from numpy import fabs, linspace #import absolute value function, fabs from numpy library, linspace for creating arrays
import matplotlib.pyplot as plt #for plotting

Xold=10 #initial guess
error=1.0e9 #some large value 
tolerance=1.0e-15

Xa=[] #empty arrays to store X values
fXa=[] #empty arrays to store f(X) values
it = 0
while error>tolerance and it < 200:
    print ("X=", Xold) 
    Xnew= ((np.sin(Xold/1.5))+Xold/1.5) #new value of X
    Xa.append(Xold) #append Xnew to Xa
    fXa.append(Xnew)    #append Y to Y
    error=fabs(Xnew-Xold) #absolute value of error
    Xold=Xnew #set Xold to Xnew for next iteration
    it = it+1
    
print ("no. of iterations= ", it)

#Plotting
plt.xlabel('X')
plt.ylabel('Xnew')
plt.plot(Xa,fXa,'*-',Xa,Xa) 
plt.legend(["Xnew=f(X)","Xnew=X"],loc="upper left")
plt.show()
```

```python
# Partial Substitutionv- Good Convergence
from numpy import fabs, linspace #import absolute value function, fabs from numpy library, linspace for creating arrays
import matplotlib.pyplot as plt #for plotting

Xold=10 #initial guess
error=1.0e9 #some large value 
tolerance=1.0e-15

alpha=2 #under-relaxation

Xa=[] #empty arrays to store X values
fXa=[] #empty arrays to store f(X) values

it=0

while error>tolerance and it<200:
    Xnew=alpha*((np.sin(Xold/1.5))+Xold/1.5)+(1.0-alpha)*Xold #new value of X
    Xa.append(Xold) #append Xnew to Xa
    fXa.append(Xnew)    #append Y to Y
    error=fabs(Xnew-Xold) #absolute value of error
    Xold=Xnew #set Xold to Xnew for next iteration
    print (Xold)
    it=it+1
print ("X=", Xold) 
if error<tolerance:
    print ("Converged reached in ", it, " iterations")
else: 
    print ("Not converged in ", it, " iterations")
#Plotting
plt.xlabel('X')
plt.ylabel('Xnew')
plt.plot(Xa,fXa,'*-',Xa,Xa) 
plt.legend(["Xnew=f(X)","Xnew=X"],loc="upper left")
plt.show()
```

```python
#Newton Raphson Method-Good Convergence

from numpy import fabs, linspace #import absolute value function, fabs from numpy library, linspace for creating arrays
import matplotlib.pyplot as plt #for plotting

Xold=10 #initial guess
error=1.0e9 #some large value 
tolerance=1.0e-15

it=1

Xa=[] #empty arrays to store X values
fXa=[] #empty arrays to store f(X) values
while error>tolerance:
    print ("X=", Xold) 
    Xnew= (Xold-((np.sin(Xold/1.5))+Xold/1.5-Xold)/((np.cos(Xold/1.5))/1.5-1/3)) #new value of X
    Xa.append(Xold) #append ew to Xa
    fXa.append(Xnew)    #append Y to Y
    error=fabs(Xnew-Xold) #absolute value of error
    Xold=Xnew #set Xold to Xnew for next iteration
    it=it+1

print ("Interations= ", it)
    
#Plotting
plt.xlabel('X')
plt.ylabel('Xnew')
plt.plot(Xa,fXa,'*-',Xa,Xa) 
plt.legend(["Xnew=f(X)","Xnew=X"],loc="lower right")
plt.show()
```
